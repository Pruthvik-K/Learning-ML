```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['town','west windsor'], axis = 'columns')
final
# drop 'town' as it not useful
# drop any dummy column from one hot encoding as it creates dummy variable trap
# if we use sklearn it knows about dummy variable trap and will drop a column implicitly
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price',axis = 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]]) #2800 sqft in robinsville
```

```python
model.predict([[3400,0,0]]) #2800 sqft in west windsor
```

```python
model.score(X,y)  # to check accuracy of the model
```
